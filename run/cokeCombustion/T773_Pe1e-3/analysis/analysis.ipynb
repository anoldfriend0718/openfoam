1. At the initial temperature, the local O2 flux near the inlet is sufficient due to the intensitive diffusive flux. Therefore, the local combustion front can increased a lot with the higher combustion front. The initial combustion temperature of the case with the initial temperature of 773K increased to 1070K, while the initial combustion temperature of the case with the initial temperature of 873K increased to about 1200K. 
Therefore, the higher initial temperature may introduce the high comnbustion temperature near the well zone in spite of the lower air injection rate
